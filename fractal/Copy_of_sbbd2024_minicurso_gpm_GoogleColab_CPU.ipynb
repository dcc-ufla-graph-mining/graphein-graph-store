# Graph Pattern Mining (GPM) concepts via NetworkX package

## Subgraphs types (induced or not-induced)
- Subgraphs are subsets of vertices and edges from an input graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# define a graph
g = nx.from_edgelist([(0,1),(1,2),(2,3),(0,3),(0,2)])
fig = plt.figure(figsize=(5,5))
fig.suptitle('input graph G', fontsize=18)
nx.draw(g, with_labels=True)

# example of induced subgraph from g
s1 = nx.subgraph(g, [0,1,2])
fig = plt.figure(figsize=(5,5))
fig.suptitle('induced subgraph', fontsize=18)
nx.draw(s1, with_labels=True)

# example of general subgraph from g
s2 = nx.edge_subgraph(g, [(0,1),(0,2)])
fig = plt.figure(figsize=(5,5))
fig.suptitle('general subgraph', fontsize=18)
nx.draw(s2, with_labels=True)
```

## Patterns
- Patterns are templates used to categorize subgraphs, composed of connectivity information plus labeling information

```python
import networkx as nx
import matplotlib.pyplot as plt

g = nx.from_edgelist([(0,1),(1,2),(2,3),(0,3),(0,2)])
fig = plt.figure(figsize=(3,3))
fig.suptitle('chordal square pattern', fontsize=18)
nx.draw(g, with_labels=False)

g = nx.from_edgelist([(0,1),(1,2),(0,3),(2,3)])
fig = plt.figure(figsize=(3,3))
fig.suptitle('square pattern', fontsize=18)
nx.draw(g, with_labels=False)

g = nx.from_edgelist([(0,1),(1,2),(0,3),(2,3)])
fig = plt.figure(figsize=(3,3))
fig.suptitle('square pattern with vertex labels', fontsize=18)
node_colors = ['red', 'red', 'blue', 'orange']
nx.draw(g, with_labels=False, node_color=node_colors)
```

## Isomorphism among subgraphs
- Notice how many subgraph isomorphisms represent the SAME subgraph

```python
import networkx as nx
from networkx.algorithms import isomorphism
import matplotlib.pyplot as plt

# define a graph (g) and a pattern (h)
g = nx.from_edgelist([(0,1),(1,2),(2,3),(0,3),(0,2)])
h = nx.from_edgelist([(0,1),(0,2),(1,2)])

fig = plt.figure(figsize=(2,2))
fig.suptitle('input graph = G', fontsize=18)
nx.draw(g, with_labels=True)

fig = plt.figure(figsize=(2,2))
fig.suptitle('pattern = H', fontsize=18)
nx.draw(h, with_labels=False)

# get all subgraphs of g isomorphic to h
for m in isomorphism.GraphMatcher(g, h).subgraph_isomorphisms_iter():
  s = nx.Graph(h)
  plt.figure(figsize=(2,2))
  nx.relabel_nodes(s, m)
  nx.draw(s, with_labels=True)
  print(m)
```

## Isomorphisms among patterns

```python
import networkx as nx
from networkx.algorithms import isomorphism
import matplotlib.pyplot as plt

# define a pattern (g)
g = nx.from_edgelist([(0,1),(1,2),(2,3),(0,3),(0,2)])
fig = plt.figure(figsize=(2,2))
fig.suptitle('pattern G', fontsize=18)
nx.draw(g, with_labels=True)

# get all automorphisms of g
for m in isomorphism.GraphMatcher(g, g).subgraph_isomorphisms_iter():
  s = nx.Graph(h)
  print(m)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

vlabeled_graph = fc.vertex_labeled_graph("fractal/data/toygraph/")

labels_of_interest = set([2,4,3])
def filter_func(s):
    for vlabel in nx.get_node_attributes(s, 'label').values():
        if vlabel not in labels_of_interest: return False
    return True

subgraphs_app = vlabeled_graph.vfractoid()
subgraphs_app = subgraphs_app.extend(4)
subgraphs_app = subgraphs_app.filter(filter_func)
subgraphs = subgraphs_app.subgraphs_networkx()

subgraphs = subgraphs.collect()[:5]
subgraphs_titles = {s:f"" for s in subgraphs}
fig = vizutil.draw_graphs_in_grid(subgraphs_titles, ncols=5, figsize=(10,2), with_labels=True, vmax=4, node_size=1000, font_size=20)
fig.tight_layout()
fig.savefig("toygraph-vertex-induced-example.pdf", bbox_inches="tight")
```

# Fractal Use-Case: Ego Networks for node similarity

## Ego-Neworks extraction

```python
citeseer = fc.vertex_labeled_graph("fractal/data/citeseer")
ego_nets = citeseer.khop_induced_subgraphs(2).collect()
ego_nets = sorted(ego_nets, key=lambda en: list(en.nodes())[0])
```

## Ego-Networks pairwise similarity computation via Graph Edit Distance (GED)

```python
import random

def node_match(n1, n2): return n1['label'] == n2['label']

ego_nets = [en for en in ego_nets if en.number_of_nodes() == 5]
random.shuffle(ego_nets)
ego_nets = ego_nets[:10]
en_map = {}
for en in ego_nets:
    en_map[en] = len(en_map)
geds = [[None] * len(ego_nets) for _ in range(len(ego_nets))]
i = 0
total = len(ego_nets) * len(ego_nets)
min_pair = None
min_d = None
max_pair = None
max_d = None
for en1 in ego_nets:
    k1 = en_map[en1]
    for en2 in ego_nets:
        k2 = en_map[en2]
        ged = nx.graph_edit_distance(en1, en2, node_match=node_match)
        geds[k1][k2] = ged
        if k1 != k2 and ged != 0 and (min_d is None or ged < min_d):
            min_d = ged
            min_pair = (en1, en2)
        if k1 != k2 and ged != 0 and (max_d is None or ged > max_d):
            max_d = ged
            max_pair = (en1, en2)
        i += 1
        print(f"[{i}/{total}] ged({en1},{en2}) = {ged}")
```

## Closest ego-networks (excluding exactly equal ones)

```python
subgraphs_titles = {en:f"Vertex={list(en.nodes())[0]}" for en in list(min_pair)}
fig = vizutil.draw_graphs_in_grid(subgraphs_titles, ncols=2, figsize=(10,3), with_labels=True, font_size=16, node_size=2000, vmax=5)
fig.tight_layout()
fig.suptitle(f"GED = {int(min_d)}", fontsize=20)
fig.savefig(f"ged-egonets-min.pdf", bbox_inches="tight")
```

## Farthest ego-networks

```python
subgraphs_titles = {en:f"Vertex={list(en.nodes())[0]}" for en in list(max_pair)}
fig = vizutil.draw_graphs_in_grid(subgraphs_titles, ncols=2, figsize=(10,3), with_labels=True, font_size=16, node_size=2000, vmax=5)
fig.tight_layout()
fig.suptitle(f"GED = {int(max_d)}", fontsize=20)
fig.savefig(f"ged-egonets-max.pdf", bbox_inches="tight")
```
