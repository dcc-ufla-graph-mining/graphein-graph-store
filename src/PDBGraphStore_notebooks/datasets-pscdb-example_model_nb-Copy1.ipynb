```python
df = pd.read_csv('structural_rearrangement_data.csv')
df.head()
```

```python
labels = pd.get_dummies(df.motion_type).values.tolist()
labels = [torch.Tensor(i) for i in labels]
```

```python
params_to_change = {
    'granularity': 'CA',
    'insertions': False,
    'keep_hets': [],
    'node_featuriser': 'meiler',
    'get_contacts_path': '',
    'pdb_dir': '../../data/pdb_files/',
    'contacts_dir': '../../data/contacts/',
    'exclude_waters': True,
    'covalent_bonds': False,
    'include_ss': True,
}
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, labels, test_size=0.15)
```

```python
pdb_store = PDBGraphStore(None)
```

```python
for i in range(len(df)):
    pdb_code = df['Free PDB'].iloc[i]
    chains = list(df['Free Chains'].iloc[i])

    params_to_change['chain_selection'] = chains
    config = ProteinGraphConfig(**params_to_change)
    
    if os.path.exists(f'../../data/pdb_files/{pdb_code}.pdb'):
        pdb_file = os.path.abspath(f'../../data/pdb_files/{pdb_code}.pdb')
    else:
        try:
            pdb_file = download_pdb(pdb_code, '../../data/pdb_files/')
        except Exception as e:
            print(e)
            print('erro 1')
            pass
    if pdb_file ==None:
        print('error')
        pass
    
    try:
        g = construct_graph(path=pdb_file, config=config)

        g_config = g.graph['config']
        g_pdb_code = g.graph['pdb_code']
        g.graph.clear()
        g.graph['config'] = g_config
        g.graph['pdb_code'] = g_pdb_code

        pdb_store.insert({pdb_code: g})

    except Exception as e:
        print(e)
        print('erro 2')
    
del g
```

```python
format_convertor = GraphFormatConvertor('nx', 'dgl',
                                        verbose = 'gnn',
                                        columns = None)
```
