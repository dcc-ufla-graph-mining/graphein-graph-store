```python
from pathlib import Path


def read_times(file_path):
    """Reads one execution time per line and returns a list of floats."""
    with open(file_path, "r") as f:
        return [float(line.strip()) for line in f if line.strip()]


def compare(file1, file2):
    times1 = read_times(file1)
    times2 = read_times(file2)

    avg1 = sum(times1) / len(times1)
    avg2 = sum(times2) / len(times2)

    lower = min(avg1, avg2)
    higher = max(avg1, avg2)

    improvement = 1 - (lower / higher)

    print(f"{Path(file1).name}")
    print(f"  Average: {avg1:.4f} s")

    print(f"{Path(file2).name}")
    print(f"  Average: {avg2:.4f} s")

    if avg1 < avg2:
        print(
            f"\nGraphein is {improvement * 100:.2f}% faster than PDBGraphStore"
        )
    elif avg2 < avg1:
        print(
            f"\nPDBGraphStore is {improvement * 100:.2f}% faster than Graphein"
        )
    else:
        print("\nBoth executions have identical average execution time.")


if __name__ == "__main__":
    compare("Graphein_notebooks/baseline_time.txt", "PDBGraphStore_notebooks/baseline_time.txt")
```
