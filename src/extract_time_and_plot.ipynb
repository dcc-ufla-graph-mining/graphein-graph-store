```python
from pathlib import Path
import numpy as np


def read_times(file_path):
    """Reads one execution time per line and returns a list of floats."""
    with open(file_path, "r") as f:
        return [float(line.strip()) for line in f if line.strip()]


def compare(file1, file2):
    times1 = read_times(file1)
    times2 = read_times(file2)

    avg1 = np.mean([times1])
    avg2 = np.mean([times2])

    stdev1 = np.std([times1])
    stdev2 = np.std([times2])

    lower = min(avg1, avg2)
    higher = max(avg1, avg2)

    improvement = 1 - (lower / higher)

    print(f"{str(Path(file1).parent).split('_')[0]}")
    print(f"  Average: {avg1:.4f} s")
    print(f'  Standard deviation: {stdev1: 4f}')

    print(f"{str(Path(file2).parent).split('_')[0]}")
    print(f"  Average: {avg2:.4f} s")
    print(f'  Standard deviation: {stdev2: 4f}')

    if avg1 < avg2:
        print(
            f"\nGraphein is {improvement * 100:.2f}% faster than PDBGraphStore\n"
        )
    elif avg2 < avg1:
        print(
            f"\nPDBGraphStore is {improvement * 100:.2f}% faster than Graphein"
        )
    else:
        print("\nBoth executions have identical average execution time.")
```

```python
dataset = 'baseline'
compare(f"Graphein_notebooks/{dataset}_time.txt", f"PDBGraphStore_notebooks/{dataset}_time.txt")
```
